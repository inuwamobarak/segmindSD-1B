## Import the necessary modules and initialize the model:

```python
from diffusers import StableDiffusionXLPipeline
import torch

# Initialize the pipeline using the pre-trained SSD-1B model:
pipe = StableDiffusionXLPipeline.from_pretrained("segmind/SSD-1B", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")

# Set the device to use (in this case, it's set to "cuda" for GPU acceleration):
pipe.to("cuda")
```

## Define your prompts.

```python
# You can change these prompts to generate different images:
prompt = "An astronaut riding a green horse"  # Your prompt here
neg_prompt = "ugly, blurry, poor quality"  # Negative prompt here
```

## Generate an image based on the provided prompts:

```python
image = pipe(prompt=prompt, negative_prompt=neg_prompt).images[0]

# You can now use the 'image' variable to work with the generated image.
```
